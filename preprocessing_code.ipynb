```python
import pandas as pd
import os 
```

```python
# 데이터 불러오기
df_final = pd.read_csv('/content/drive/MyDrive/추천시스템_컨퍼런스/final_data_1236.csv')
df_final.drop('Unnamed: 0', axis = 1, inplace = True)
```

# 키워드 뽑기 및 평점 및 리뷰 개수 구하기

```python
# 키워드 뽑기
def get_keyword(x):
    return x.split(' ')[-1]

# 키워드 축소
def get_keyword2(k):
    hotel_lst = ['리조트', '콘도', '객실']
    pension_lst = ['전원주택', '펜션', '스위트', '주택', '저택', '통나무집', '숙소', '오두막', '로프트', '집', '타운하우스']
    camping_lst = ['캠핑카', '텐트', '야영장', '돔하우스']

    if k in hotel_lst:
        kw2 = '호텔 * 리조트'
    elif k in pension_lst:
        kw2 = '펜션 * 빌라'
    elif k in camping_lst:
        kw2 = '캠핑'
    else:
        kw2 = k
    return kw2
    
# 평점 뽑기
def get_score(x):
    if x == '신규 숙소':
        return 'x'
    else:
        temp =  x.split(', 후기 ')[0]
        score = float(temp[3:].split('점')[0])
        return score
# 후기 개수 뽑기
def get_review(x):
    if x == '신규 숙소':
        return 0
    else:
        return x.split(', 후기 ')[1][:-1]

df_final['keyword'] = df_final['숙소이름'].apply(get_keyword)
df_final['keyword2'] = df_final['keyword'].apply(get_keyword2)
df_final['평점'] = df_final['리뷰평점 및 개수'].apply(get_score)
df_final['후기개수'] = df_final['리뷰평점 및 개수'].apply(get_review)

# keyword2 원핫 인코딩
encoded_keyword = pd.get_dummies(df_final['keyword2'])
encoded_df2 = pd.concat([df_final, encoded_keyword], axis = 1)
encoded_df2



```

# 중복 데이터 처리

```python
# 중복데이터 삭제
dup_del_df = encoded_df2.drop_duplicates(['한줄소개', '숙소이름', '평점', '후기개수'], keep = 'first', ignore_index = True)
# 컬럼 명 바꾸기
df_data_preprocessed = dup_del_df.rename(
    columns={
        '이미지 인덱스' : 'image_index',
        '이미지주소': 'image_url',
        '평점': 'rating',
        '후기개수': 'comment',
        '개인실': 'single',
        '다인실': 'multi',
        '별채': 'outhouse',
        '아파트': 'apartment',
        '캠핑': 'camping',
        '펜션 * 빌라': 'pension',
        '호텔 * 리조트': 'hotel',
    })
df_data_preprocessed.set_index('image_index', inplace = True)
df_data_preprocessed
```
